```python
import numpy as np
import matplotlib.pyplot as plt
```

Создаем список из 100 значений от 0 до 2

```python
# Создаем список из 100 значений от 0 до 2
x = np.linspace(0, 2, 100)
type(x)
```

Объявляем пустое поле для графика (figure) 500 на 300 px

```python
x = np.linspace(0, 2, 100)
# Создаем пустое поле
plt.figure(figsize=(5, 3), layout='constrained')
```

Строим линейный график

```python
x = np.linspace(0, 2, 100)
plt.figure(figsize=(5, 3), layout='constrained')
plt.plot(x, x, label='linear')
```

Добавляем квадратичный график и легенду

```python
x = np.linspace(0, 2, 100)

plt.figure(figsize=(5, 3), layout='constrained')
plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic') 

plt.legend()
plt.show()
```

Добавляем заголовок и подписи осей

```python
x = np.linspace(0, 2, 100)

plt.figure(figsize=(5, 3), layout='constrained')
plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic') 
plt.plot(x, x**3, label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")


plt.legend()
plt.show()
```

Настраиваем риски (ticks) для оси X,  изменяем размеры фигуры на более подходящие

```python
x = np.linspace(0, 2, 100)

plt.figure(figsize=(8, 3), layout='constrained')
plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic') 
plt.plot(x, x**3, label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.xticks(np.arange(0, 2.1, 0.1))
plt.legend()
plt.show()
```

Настраиваем риски для оси Y и выводим сетку (grid)

```python
x = np.linspace(0, 2, 100)

plt.figure(figsize=(8, 3), layout='constrained')
plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic') 
plt.plot(x, x**3, label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.xticks(np.arange(0, 2.1, 0.1))
plt.yticks(np.arange(0, 9, 1))
plt.grid(True)
plt.legend()
plt.show()
```

```python
x = np.linspace(0, 2, 100)

# plt.style.use("dark_background")
# print(plt.style.available)

plt.figure(figsize=(8, 3), layout='constrained')
plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic') 
plt.plot(x, x**3, label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.xticks(np.arange(0, 2.1, 0.1))
plt.yticks(np.arange(0, 9, 1))
plt.grid(True)
plt.legend(loc="upper left")
plt.margins(x=0.02, y=0.1)
plt.show()
```

```python
xmin = 0
xmax = 100


x = np.linspace(xmin, xmax, 1000)

# plt.style.use("dark_background")
# print(plt.style.available)

plt.figure(figsize=(6, 8), layout='constrained')
plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic') 
plt.plot(x, x**3, label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Сравнение квадратической и кубической зависимостей")
plt.xticks(np.arange(xmin, xmax+1, 5))
plt.yticks(np.arange(0, 1000001, 100000))
plt.grid(True)
plt.legend(loc="upper left")
plt.margins(x=0.02, y=0.1)
plt.show()
```
