```python
import numpy as np
import matplotlib.pyplot as plt

# Начальная сумма
value = 1000 

# На сколько лет
n = 30

# Годовой процент
percent = 7



years = np.linspace(0, n, n+1)

simple = value * (1 + percent/100 * years)
compound = value * (1 + percent/100)**years

plt.figure(figsize=(6, 4), layout='constrained')
plt.plot(years, simple, ".-", label='Простой процент')
plt.plot(years, compound, ".-", label='Сложный процент') 

plt.xticks(np.arange(0, 31, 2))
plt.yticks(np.arange(value, compound[-1]+1000, 1000))

plt.xlabel("Годы")

plt.grid(True)

plt.title("Приращение средств")

plt.legend()
plt.show()
```
