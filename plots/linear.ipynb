```python
# Slope
x1 = 1
y1 = 5
x2 = 2
y2 = 8

m = (y2 - y1) / (x2 - x1)

print("slope = ", m)

b = y1 - m*x1
print("intercept = ", b)


# Полное уравнение
print("y = ", m, "x + ", b)
```
