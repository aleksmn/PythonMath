```python
import numpy as np
import matplotlib.pyplot as plt

E = 3  # Напряжение в вольтах
R_values = np.linspace(1, 20, 100)  # Значения сопротивления от 1 до 20 Ом

# Рассчитаем ток по закону Ома
I = E / R_values

# Создадим график
plt.figure(figsize=(10, 6))
plt.plot(R_values, I, label=f'I = {E} / R', color='blue')
plt.title("Закон Ома: ток к сопротивлению")
plt.xlabel("Сопротивление (R) в Омах")
plt.ylabel("Ток (I) в Амперах")
plt.grid()

plt.legend()
plt.xlim(0, 20)
plt.ylim(0, E / 1)  # Максимальный ток, когда минимальное сопротивление
plt.show()


# Получилась линия, которая стремится к осям, но никогда не касается их. Это гипербола.
```
