```python
import numpy as np
import matplotlib.pyplot as plt

U = 3  # Напряжение в вольтах
R_values = np.linspace(1, 20, 100)  # Значения сопротивления от 1 до 20 Ом

# Рассчитаем ток по закону Ома
I = U / R_values

# Создадим график
plt.figure(figsize=(10, 6))
plt.plot(R_values, I, label=f'I = U / R, (E={U}V)', color='blue')
plt.title("Закон Ома: ток к сопротивлению")
plt.xlabel("Сопротивление (R) в Омах")
plt.ylabel("Ток (I) в Амперах")

# Найдем ток для карманного фонарика
R1 = 4
plt.plot(R1, U/R1, "ro", label=f'I = {U/R1}А')

plt.grid()

plt.legend()
plt.xlim(0, 20)
plt.ylim(0, U / 1)  # Максимальный ток, когда минимальное сопротивление
plt.show()


# Получилась линия, которая стремится к осям, но никогда не касается их. Это гипербола.
```
